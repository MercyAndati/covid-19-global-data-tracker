```python
import pandas as pd
# Load the data
df=pd.read_csv("owid-covid-data.csv")
# Print column names
print(df.columns)
# Preview the first 5 rows
print(df.head())
# Filter countries of interest
countries=['Kenya', 'United States', 'India','Brazil','South Africa', 'Spain']
df=df[df['location'].isin(countries)]
#sum of null values
print(df.isnull().sum())
# Drop rows with missing dates or other critical values
df=df.dropna(subset=['date','location','total_cases','total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations'])
# Handle missing numeric values: fill with interpolation or a specific value
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].interpolate() 
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(df['total_cases'],df['location'])
plt.xlabel('total cases')
plt.ylabel('country')  
plt.title('total cases over time for selected countries.')
plt.show()
```

```python
plt.plot(df['total_deaths_per_million'],df['date'])
plt.xlabel('total death/million')
plt.ylabel('date')
plt.title('total deaths over time.')
plt.show()
```

```python
import matplotlib.pyplot as plt
top_countries = df.groupby('location')['total_cases'].max().sort_values(ascending=False).head(10)
plt.bar(top_countries.index ,top_countries)
plt.xlabel('countries')
plt.ylabel('total cases')
plt.xticks(rotation=45)
plt.title('top countries by total cases')
plt.show()
```

```python
plt.bar(df['location'],df['new_cases'])
plt.xlabel('country')
plt.xticks(rotation=45)
plt.ylabel('new cases')
plt.title('daily new cases between countries')
plt.show()
```

```python
death_rate = df['total_deaths'] / df['total_cases']
print(death_rate)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant numeric columns for correlation
cols = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 
        'total_cases_per_million', 'total_deaths_per_million', 
        'total_vaccinations', 'population']

# Compute correlation matrix
corr = df[cols].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('COVID-19 Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='total_vaccinations', hue='location')

plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.title('Cumulative Vaccinations Over Time for Selected Countries')
plt.xticks(rotation=45)  # Rotate dates for better readability
plt.tight_layout()       # Adjust layout to prevent clipping
plt.show()
```

```python
# Example for one country (e.g., Kenya)
latest_kenya = df[df['location'] == 'Kenya'].sort_values('date').iloc[-1]
vaccinated = latest_kenya['people_vaccinated']
unvaccinated = latest_kenya['population'] - vaccinated

sizes = [vaccinated, unvaccinated]
labels = ['Vaccinated', 'Unvaccinated']

plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Vaccination Status in Kenya (Latest)')
plt.show()
```
